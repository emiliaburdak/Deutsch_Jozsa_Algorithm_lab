```python
def oracle(qc, n):
    for i in range(n-1):
        qc.cx(i, n+1)
        qc.cx(i+1, n+1)
        qc.x(n+1)

    temp_result = list(range(n, 2*n - 1))
    ancilla = 2*n - 1
    qc.mcx(temp_result, ancilla)

    for i in reversed(range(n-1)):
        qc.x(n+1)
        qc.cx(i+1, n+1)
        qc.cx(i, n+1)

    return qc
```
